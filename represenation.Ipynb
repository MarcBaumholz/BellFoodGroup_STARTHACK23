```python
import pandas as pd
from matplotlib import pyplot as plt

df5 = pd.read_csv('Data/energy_Feldhof.csv')
df5 = df5.set_index(pd.to_datetime(df5['ts']))
df5 = df5.resample('15T').mean()
```
